```python
#Draw Chart
import matplotlib.pyplot as plt
```

```python
#Draw Chart
x = [1,2,3,4,5]
y = [10,16,20,30,25]

x1 = [1,2,3,4,5]
y1 = [12,14,22,34,20]

plt.plot(x,y,'bo-',label = "Data1")
plt.plot(x1,y1,'ro-',label = "Data2")
plt.legend()
plt.show()
```

```python
#Draw Chart
x = [1,2,3,4,5]
y = [10,16,20,30,25]

x1 = [1,2,3,4,5]
y1 = [12,14,22,34,20]

plt.plot(x,y,'bo-',label = "Data1")
plt.plot(x1,y1,marker  = 'o',color='r',linestyle ='-', label = "Data2")
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Draw Chart
categories = ['A','B','C','D','E']
values = [15,10,25,12,18]

plt.bar(categories,values,color = 'g',alpha = 0.6)
plt.xlabel('Column')
plt.ylabel('Value')
plt.show()
```

```python
#Pie Chart
labels = ['A','B','C','D']
sizes = [15,30,45,20]
colors = ['r','green','blue','yellow']

plt.pie(sizes,labels = labels,colors = colors,autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Scatter
x1 = [1,2,3,4,5]
y1 = [12,14,22,34,20]

plt.scatter(x1,y1,marker  = 'o',color='r',linestyle ='-', label = "Data2")
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Box Chart
x = [1,2,3,4,5,7,9,10,30,20]

plt.boxplot(x)
plt.title('Box Chart')
plt.ylabel('Value')
plt.show()
```
